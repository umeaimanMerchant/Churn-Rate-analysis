# Hypothesis Testing


**Hypotheses:**

$H_0$: There is no difference in average number of drives between drivers who use iPhone devices and drivers who use Androids.

$H_A$: There is a difference in average number of drives between drivers who use iPhone devices and drivers who use Androids.

```python
import scipy.stats as stats
```

```python
sign_level = 0.05

data_and = df_not_null[df_not_null['device']=='Android']['drives']
data_iPh = df_not_null[df_not_null['device']=='iPhone']['drives']

stats.ttest_ind(a= data_and, b = data_iPh)
```

Pvalues > sign_level

H0 is true- There is no difference in average number of drives between drivers who use iPhone devices and drivers who use Androids.

### **Task 3b. Determine whether assumptions have been met**

The following are the assumptions for logistic regression:

* Independent observations (This refers to how the data was collected.)

* No extreme outliers

* Little to no multicollinearity among X predictors

* Linear relationship between X and the **logit** of y

For the first assumption, you can assume that observations are independent for this project.

The second assumption has already been addressed.

The last assumption will be verified after modeling.

**Note:** In practice, modeling assumptions are often violated, and depending on the specifics of your use case and the severity of the violation, it might not affect your model much at all or it will result in a failed model.

**Check Collinearity in X**

```python
data['percent_sessions_in_last_month'] = data['sessions'] / data['total_sessions']
data['km_per_hr'] = data['driven_km_drives'] * 60 / data['duration_minutes_drives']
data['percent_nav_fav'] = (data['total_navigations_fav1']+data['total_navigations_fav2'])/data['total_sessions']
```

```python
data.info()
```

```python
# 1. Isolate X variables
X = data.drop(columns=['label', 'label2', 'device'])

# 2. Isolate y variable
y = data['label2']

# 3. Split into train and test sets
X_tr, X_test, y_tr, y_test = train_test_split(X, y, stratify=y,
                                              test_size=0.2, random_state=42)

# 4. Split into train and validate sets
X_train, X_val, y_train, y_val = train_test_split(X_tr, y_tr, stratify=y_tr,
                                                  test_size=0.25, random_state=42)
```

```python
rf = RandomForestClassifier()

cv_para = {'max_depth': [3,5, None],
          'min_samples_leaf' : [2],
          'min_samples_split': [2],
          'max_features':[1.0],
          'n_estimators' :[300],
          'max_samples': [1.0]}

scoring= ['accuracy', 'recall', 'precision', 'f1']
```

```python
rf_cv = GridSearchCV(rf, cv_para, scoring= scoring,cv=4, refit='f1')
```

```python
rf_cv.fit(X_train, y_train)
```
